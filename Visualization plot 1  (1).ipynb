```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv('Downloads\Data_Science_Topics_Survey.csv')
df.head() 
```

```python
df1 = df.drop(['Timestamp'],axis=1)
df1.head()
```

```python
df1=df1.rename(columns={"What's your level of interest for the following areas of Data Science? [Data Visualization]":"Data Vizualization",
                        "What's your level of interest for the following areas of Data Science? [Machine Learning]":"Machine Learning",
                        "What's your level of interest for the following areas of Data Science? [Data Analysis / Statistics]":"Data Analysis/Statistics",
                        "What's your level of interest for the following areas of Data Science? [Big Data (Spark / Hadoop)]":"Big Data(Spark/Hadoop)",
                        "What's your level of interest for the following areas of Data Science? [Data Journalism]":"Data Journalism",
                        "What's your level of interest for the following areas of Data Science? [Deep Learning]":"Deep Learning"})
```

```python
df1=df1.apply(pd.Series.value_counts)
```

```python
df1.sort_index(axis=1,inplace=True)
df1.transpose()
```

```python
import matplotlib as mp
import matplotlib.pyplot as plt 
```

```python
df1.sort_index(ascending=False,axis=0, inplace=True)
df1.transpose()

```

```python
#sort in the desecnding order of values in very interesting column 
df1.sort_values(by='Very interested',axis=1, ascending=False, inplace=True)
df2 = df1.transpose()
df2
```

```python
#convert the number values of column into percentages
df2['Very interested'] = (df2['Very interested']/2233)*100
df2['Somewhat interested'] = (df2['Somewhat interested']/2233)*100
df2['Not interested'] = (df2['Not interested']/2233)*100
df2
```

```python
#round to two decimal places 
np.round(df2,decimals=2)
```

```python
#plot the values in a bar plot key note: the legend value is true by default 

ax=df2.plot(kind="bar",width=0.8,figsize=(20,8),color=['#5cb85c', '#5bc0de', '#d9534f'])
plt.title("Percentage of Respondent's Interest in Data Science Areas",fontsize=16)
plt.legend(fontsize=14)
plt.xticks(fontsize=14)
plt.yticks([])
plt.grid(False)
ax.set_facecolor('white')
for p in ax.patches:
             ax.annotate("%.2f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()),
                 ha='center', va='center', fontsize=11, color='gray', xytext=(0, 20),
                 textcoords='offset points')
_ = ax.set_ylim(0,120)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
#plt.tight_layout()
plt.show()
```
